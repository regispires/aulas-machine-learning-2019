## Linha reta de uma regressão linear simples



$$ y = w_0 + w_1 \cdot x $$

```python
# $$ mean(x) = \frac { \sum_{i=1}^n { x_i } } { count(x) } $$
```

## Média

$$ mean(x) = \frac { \sum_{i=1}^n { x_i } } { count(x) } $$

```python
x = np.array([ 1, 2, 4, 3, 5 ])
y = np.array([ 1, 3, 3, 2, 5 ])
```

```python
mean_x = np.mean(x)
mean_x
```

```python
mean_y = np.mean(y)
mean_y
```

```python
# w_1 = \frac { \sum_{i=1}^n ((x_i - mean(x)) \cdot (y_i - mean(y))) } 
# { \sum_{i=1}^n (x_i - mean(x))^2 }
```

# Cálculo dos coeficientes

$$ w_1 = \frac { \sum_{i=1}^n ((x_i - mean(x)) \cdot (y_i - mean(y))) } { \sum_{i=1}^n (x_i - mean(x))^2 } $$

```python
n = x.shape[0]
n
```

```python
x - mean_x
```

```python
y - mean_y
```

```python
sum1 = np.sum( (x - mean_x) * (y - mean_y) )
sum2 = np.sum( (x - mean_x) ** 2 )
w1 = sum1 / sum2
w1
```

$$ w_0 = mean(y) - w_1 \cdot mean(x) $$

```python
w0 = mean_y - w1 * mean_x
w0
```

## Predição

```python
def predict(x):
    return w0 + w1 * x
```

```python
y_pred = predict(x)
y_pred
```

```python
def mse(y, y_pred):
    return np.sum(( y - y_pred ) ** 2)
```

```python
mse(y, y_pred)
```

```python
def rmse(y, y_pred):
    return mse(y, y_pred) ** 0.5
```

```python
rmse(y, y_pred)
```

```python
class RegressaoLinearSimples:

    def fit(self, x, y):
        mean_x = np.mean(x)
        mean_y = np.mean(y)
        sum1 = np.sum( (x - mean_x) * (y - mean_y) )
        sum2 = np.sum( (x - mean_x) ** 2 )
        self.w1 = sum1 / sum2
        self.w0 = mean_y - self.w1 * mean_x

    def predict(self, x):
        return self.w0 + self.w1 * x
```

```python
model = RegressaoLinearSimples()
```

```python
model.fit(x, y)
```

```python
y_pred = model.predict(x)
```
