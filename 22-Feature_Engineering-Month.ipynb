```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = np.arange(0, 12, step=0.1)
sin = np.sin((months)*(2.*np.pi/12))
cos = np.cos((months)*(2.*np.pi/12))
```

```python
months
```

```python
plt.scatter(months, sin, marker='.');
```

```python
plt.scatter(months, cos, marker='.');
```

```python
plt.scatter(cos, sin, marker='.')
plt.scatter(cos[0], sin[0], marker='.', c='black')
plt.scatter(cos[-1], sin[-1], marker='.', c='red')
plt.axes().set_aspect('equal')
```
